```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons

import tensorflow
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
```

```python
X , y  = make_moons(200 , noise = 0.2 , random_state = 23)
```

```python
import  matplotlib.pyplot as plt
plt.scatter(X[: , 0] , X[: , 1] , c = y)
```

```python
model1 = Sequential()

model1.add(Dense(128 , input_dim = 2 , activation = "relu"))
model1.add(Dense(128 , activation = "relu"))
model1.add(Dense(1 , activation = "sigmoid"))
```

```python
model1.summary()
```

```python
adam = Adam(learning_rate=0.01)
model1.compile(loss = "binary_crossentropy" ,optimizer=adam , metrics= ["accuracy"] )
```

```python
history1 = model1.fit(X , y , epochs = 200 ,validation_split = 0.2 ,  verbose = 0)
```

```python
plt.plot(history1.history["loss"])
plt.plot(history1.history["val_loss"])
```

```python
model2 = Sequential()

model2.add(Dense(128 , input_dim = 2 , activation= "relu"))
model2.add(Dropout(0.3))
model2.add(Dense(128 , activation = "relu"))
model2.add(Dropout(0.3))
model2.add(Dense(1 , activation="sigmoid"))

```

```python
model2.summary()
```

```python
adam = Adam(learning_rate=0.01)
model2.compile(loss = "binary_crossentropy" , optimizer= adam , metrics=["accuracy"])
```

```python
history2 = model2.fit(X , y , validation_split = 0.2 , epochs = 200 , verbose=0)
```

```python
plt.plot(history2.history["loss"])
plt.plot(history2.history["val_loss"])
```
