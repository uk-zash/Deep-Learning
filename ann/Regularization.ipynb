```python
import numpy as np 
import pandas as pd
from sklearn.datasets import make_moons

import tensorflow
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
```

```python
X , y = make_moons(100 , noise = 0.25 , random_state = 32)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[: , 0] , X[: , 1] , c=y)
plt.show()
```

```python
X.shape
```

```python
model = Sequential()

model.add(Dense(128 , input_dim=2 , activation = "relu"))
model.add(Dense(128 , activation = "relu"))
model.add(Dense(1 , activation = "sigmoid"))

model.summary()
```

```python
adam = Adam(learning_rate = 0.01)
model.compile(loss = "binary_crossentropy" , optimizer=adam , metrics = ["accuracy"])

```

```python
history1 = model.fit(X , y , epochs= 2000 , validation_split = 0.2 , verbose = 0)
```

```python
plt.plot(history1.history["loss"])
plt.plot(history1.history["val_loss"])
plt.show()
```

```python
model1 = Sequential()

model1.add(Dense(128 , input_dim = 2 , activation = "relu" , kernel_regularizer = tensorflow.keras.regularizers.l1(0.03)))
model1.add(Dense(128 ,  activation = "relu" , kernel_regularizer = tensorflow.keras.regularizers.l1(0.03)))
model1.add(Dense(1 ,  activation = "sigmoid"))
```

```python
model1.summary()
```

```python
adam = Adam(learning_rate=0.01)
model1.compile(loss = "binary_crossentropy" , optimizer=adam , metrics = ["accuracy"])
```

```python

history2 = model1.fit(X , y , epochs = 2000 , validation_split = 0.2 , verbose = 0)
```

```python
plt.plot(history2.history["loss"])
plt.plot(history2.history["val_loss"])
```
